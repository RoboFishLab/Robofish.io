## Calibration for RoboFish Pump

### Material:
- 12 V Peristaltic Pump
- ESP32 WROOM32 Microcontroller
- TB6612 DC/Stepper Driver
- 2 mm Diameter Silikon tube

### Proceedure
Pump was run with 12V for 30s with 40%, 50%, 80% and 100% PWM with RO water.  
The mass of the water was calculated to acquire the rate in ml/s.


```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
```

```python
# x = np.array([0.4, 0.5, 0.8, 1.0]) # %
x = np.array([40,50,80,100])
y = np.array([0.061, 0.077, 0.173, 0.23]) # ml/s

a, b = np.polyfit(x, y, 1)
print('Rate of change: ', a)

fig,ax = plt.subplots(figsize=(10,5))
plt.plot(x,y)
plt.grid(alpha=0.4)
plt.show()
```
