## Calibration for RoboFish Pump

### Material:
- 12 V Peristaltic Pump
- ESP32 WROOM32 Microcontroller
- TB6612 DC/Stepper Driver
- 2 mm Diameter Silikon tube

### Proceedure
Pump was run with 12V for 30s with 40%, 50%, 80% and 100% PWM with RO water.  
The mass of the water was calculated to acquire the rate in ml/s.


```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
```

Measurements:

    Empty vial:    21.82 mg  
    50%:           23.97 mg   
    60%:           24.93 mg  
    70%:           25.75 mg
    80%:           26.44 mg
    90%:           27.56 mg
    100%:          28.35 mg

```python
x = np.array([50,60,70,80,90,100])
y = (np.array([23.97,24.93,25.75,26.44,27.56,28.35]) - 21.82) / 30

a, b = np.polyfit(x, y, 1)
print('Rate of change: ', a)

fig,ax = plt.subplots(figsize=(10,5))
ax.plot(x,y)
ax.grid(alpha=0.4)
ax.xaxis.set_tick_params(labelsize=14)
ax.yaxis.set_tick_params(labelsize=14)
plt.ylabel('Output (ml/s)', fontsize=16)
plt.xlabel('Duty Cycle (%)', fontsize=16)
plt.show()
```
